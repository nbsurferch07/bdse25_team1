## Import modules

```python
import numpy as np
import pandas as pd
import os, sys
```

## Load data 

```python
df = pd.read_csv('dataset.csv', engine='python')
df.index += 1
```

```python
df
```

## Data processing

```python
new_df = df.copy()
new_df
```

```python
# select needed rows 
con1 = ~df["交易標的"].isin(["車位","土地"]) # remove garage & land
con2 = df["主要用途"].isin(["住家用"]) # select type of building
con3 = df['樓層'].notna() # remove NaN
con4 = df['單價元平方公尺'].notna() # remove NaN
con5 = df["交易年"].isin(list(range(100, 111))) # year range 100-111 
con6 = df["交易月"].isin(list(range(1, 13))) # month range 1-12 

new_df = new_df.loc[con1 & con2 & con3 & con4 & con5 & con6]

# filter floor 1~20
new_df = new_df.groupby("樓層").filter(lambda grp: len(grp) > 1000) 
new_df
```

```python
new_df = new_df[["城市代碼", "鄉鎮市區", "建物型態", "交易年", "交易月", "樓層", "建物現況格局-房", "建物現況格局-廳", "建物現況格局-衛", "建物移轉總面積平方公尺", "每坪價格", "價格分類"]]
new_df
```

```python
new_df.to_csv('dataset_drawing.csv', index = False)
```
