```python
# 輸出黑色像素
import numpy as np
import cv2

# load image
img = cv2.imread("./raw/001.jpg")


# Convert BGR to HSV
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# define range of black color in HSV
lower_val = np.array([0,0,0])
upper_val = np.array([0,255,150])

# Threshold the HSV image to get only black colors
mask = cv2.inRange(hsv, lower_val, upper_val)

# invert mask to get black symbols on white background
mask_inv = cv2.bitwise_not(mask)

# display image
cv2.imshow("Mask", mask_inv)
cv2.imshow("Img", img)

# save image
cv2.imwrite('rm001.jpg',mask_inv)

cv2.waitKey(0)
cv2.destroyAllWindows()

# 參考: https://stackoverflow.com/questions/54978271/opencv-plot-black-pixels-on-a-image

# 短期
# 邊緣缺角,因為jpeg壓縮
# 批量處理

# 中期
# 彩色圖片(可以轉灰階,去噪)
# 文字辨識,分割
# 家具辨識,分割
```

# 批量處理

```python
# 輸出黑色像素
import numpy as np
import cv2
import os

# define range of black color in HSV
lower_val = np.array([0,0,0])
upper_val = np.array([0,255,150])

# load image in file
file_root = "./raw" 
file_list = os.listdir(file_root)
watermark_out = "./rmWatermark/"
Noising_out = "./Denoising/"

for img_name in file_list:
    if img_name.endswith('.jpg'):
        img_path =file_root +'/'+img_name
        img = cv2.imread(img_path)
        hsv= cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
        mask = cv2.inRange(hsv, lower_val, upper_val)
        mask_inv = cv2.bitwise_not(mask)
        out_name = img_name.split('.')[0]
        
#  檔名前加上rmw, 代表已完成移除浮水印的處理
        save_path = watermark_out + 'rmw' + out_name + '.jpg'
        cv2.imwrite(save_path,mask_inv)

#降噪＆平滑 
# 移除躁點 檔名前加上Ｎ, 代表已完成移除噪點
#         mean = cv2.medianBlur(mask_inv, 3)
#         save_path = Noising_out + 'N' + out_name + '.jpg'
#         cv2.imwrite(save_path,mean)
        
        mean = cv2.blur(mask_inv, (3,3))
        save_path = Noising_out + 'N' + out_name + '.jpg'
        cv2.imwrite(save_path,mean)   
        
# 參考:　https://blog.csdn.net/qq_35872942/article/details/107774134
# 參考:　https://blog.csdn.net/qq_36758914/article/details/104027340


# 中期
# 彩色圖片(可以轉灰階,去噪)
# 文字辨識,分割
# 家具辨識,分割
```
